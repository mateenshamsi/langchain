## Sentence Transformers on Hugging Face

Sentence Transformers are powerful models for generating dense vector representations (embeddings) of sentences and text. On Hugging Face, you can find a wide variety of pre-trained Sentence Transformer models suitable for tasks like semantic search, clustering, and classification.

To use Sentence Transformers with Hugging Face, you typically load a model using the `sentence-transformers` library or integrate it with frameworks like LangChain for advanced workflows.

**Example Hugging Face models:**
- `sentence-transformers/all-MiniLM-L6-v2`
- `sentence-transformers/paraphrase-MiniLM-L3-v2`
- `sentence-transformers/msmarco-MiniLM-L-12-v3`

Visit [Hugging Face Sentence Transformers](https://huggingface.co/sentence-transformers) for more models and documentation.

```python
from langchain_huggingface import HuggingFaceEmbeddings
HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2",
    )
```

```python
text="this is a test sentence for embedding"
query_result = HuggingFaceEmbeddings().embed_query(text)
```
